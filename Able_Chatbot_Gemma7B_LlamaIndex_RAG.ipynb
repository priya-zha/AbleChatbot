```python
import logging
import sys

# logging.basicConfig(stream=sys.stdout, level=logging.INFO)
# logging.getLogger().addHandler(logging.StreamHandler(stream=sys.stdout))
# Import LlamaIndex modules
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core import Settings
from llama_index.embeddings.fastembed import FastEmbedEmbedding
from llama_index.core import PromptTemplate
import torch

# loading documents from the "/content/Able" directory for indexing

able_file = SimpleDirectoryReader("/content/Able").load_data()
```

```python
# initializing the embedding model using FastEmbed
fastEmbedModel = FastEmbedEmbedding(model_name="BAAI/bge-small-en-v1.5")
Settings.embed_model = fastEmbedModel #  setting embeddingmodel for LlamaIndex
Settings.chunk_size = 512  #  chunk size for splitting documents
```

```python
#  query prompt template to wrap user queries
sysPrompt = """You are a helpful AI assistant. You should answer user questions based on the provided documents. If no relevant information is found, provide a reasonable response based on your general knowledge.
You can also engage in small talk, greetings, and casual conversation."""


# This will wrap the default prompts that are internal to llama-index
Queryprompt = PromptTemplate("<|USER|>{query_str}<|ASSISTANT|>")


```

```python
#hugging face api token
from huggingface_hub import  notebook_login
notebook_login()
```

```python

# initializingg HuggingFace-based LLM with Gemma7B model
llm = HuggingFaceLLM(
    context_window=8192,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.7, "do_sample": False},
    system_prompt=sysPrompt,
    query_wrapper_prompt=Queryprompt,
    tokenizer_name="google/gemma-7b-it",
    model_name="google/gemma-7b-it",
    device_map="auto",
    #stopping_ids=[50278, 50279, 50277, 1, 0],
    tokenizer_kwargs={"max_length": 4096},
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16}
)

#  LLM and chunk size in LlamaIndex settings

Settings.llm = llm
Settings.chunk_size = 512
```

```python
indexing = VectorStoreIndex.from_documents(able_file)  #Creating a vector index from loaded documents
```

```python

queryEng = indexing.as_query_engine()  # converting index into a query enginee for answering questions

# def predict(input, history):
#   response = query_engine.query(input)
#   return str(response)
```

```python
# function to process user queries
def answerPrediction(inputQuestion, *args):
    answer = queryEng.query(inputQuestion)
    return str(answer)

```

#This part shows the working of the chatbot. It's created with python CLI commands. You can view the results here. Additionally, I have also used gradio to create a chatbot interface present below.

```python
# Loop to continuously ask for questions and provide answers
while True:
    userquestion = input("\nPlease enter your question (or type 'exit' to quit): ")

    if userquestion.lower() == 'exit':
        print("Exiting the program.")
        break

    answer = answerPrediction(userquestion)
    print("Answer: ", answer)
```
